**Basic Exploratory Data Analysis and Filtering**

```python
ge.head()
#ge.describe()
```

```python
countryList=["Sweden","Finland","Norway","Iceland","Denmark"]
#creating a list for major Nordic countries
```

```python
nordic_data=(ge[(ge["country"].isin(countryList)) & (ge['year'] >= 1990) & (ge['year'] <= 2018)])
#extracting the data for major Nordic countries by using a list and further filtering it for the duration between the year 1990 till 2018
```

```python
nordic_data
```

```python
nordic_data.info()
#displaying basic information regarding the datatypes
```

```python
new_nordic_data=nordic_data[['iso_code','country','year','co2','total_ghg','methane','nitrous_oxide']]
#further filtering out the extra columns from nordic_data
```

```python
new_nordic_data
#displaying filtered data
```

```python
new_nordic_data.info()
#checking data types of the new filtered data
```

```python
sweden=(new_nordic_data[(new_nordic_data["country"] == 'Sweden')])
#storing Sweden's gas emissions data in its own object
```

```python
norway=(new_nordic_data[(new_nordic_data["country"] == 'Norway')])
#storing Norway's gas emissions data in its own object
```

```python
denmark=(new_nordic_data[(new_nordic_data["country"] == 'Denmark')])
#storing Denmark's gas emissions data in its own object
```

**Visualizations for Gas Emissions**

```python
plt.figure(figsize=(15,8))
plt.plot(sweden['year'],sweden['co2'],color='green',label='Sweden',marker='d')
plt.plot(denmark['year'],denmark['co2'],color='red',label='Denmark', marker='d')
plt.plot(iceland['year'],iceland['co2'],color='magenta',label='Iceland', marker='d')
plt.plot(norway['year'],norway['co2'],color='blue',label='Norway', marker='d')
plt.plot(finland['year'],finland['co2'],color='orange',label='Finland', marker='d')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Carbon Dioxide mT",fontsize=15)
plt.title('Total CO2 Emissions Over Time',fontsize=15)
plt.legend()
plt.savefig('Total CO2 Emissions Over Time.png', dpi=300, bbox_inches='tight')
plt.show()
#for Carbon Dioxide CO2
```

```python
plt.figure(figsize=(15,8))
plt.plot(sweden['year'],sweden['methane'],color='green',label='Sweden',marker='d')
plt.plot(denmark['year'],denmark['methane'],color='red',label='Denmark', marker='d')
plt.plot(iceland['year'],iceland['methane'],color='magenta',label='Icleand', marker='d')
plt.plot(norway['year'],norway['methane'],color='blue',label='Norway', marker='d')
plt.plot(finland['year'],finland['methane'],color='orange',label='Finland', marker='d')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Methane mT",fontsize=15)
plt.title('Total CH4 Emissions Over Time',fontsize=15)
plt.legend()
plt.savefig('Total CH4 Emissions Over Time.png', dpi=300, bbox_inches='tight')
plt.show()
#For Methane CH4
```

```python
plt.figure(figsize=(15,8))
plt.plot(sweden['year'],sweden['nitrous_oxide'],color='green',label='Sweden',marker='d')
plt.plot(denmark['year'],denmark['nitrous_oxide'],color='red',label='Denmark', marker='d')
plt.plot(iceland['year'],iceland['nitrous_oxide'],color='magenta',label='Icleand', marker='d')
plt.plot(norway['year'],norway['nitrous_oxide'],color='blue',label='Norway', marker='d')
plt.plot(finland['year'],finland['nitrous_oxide'],color='orange',label='Finland', marker='d')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Nitrous Oxide mT",fontsize=15)
plt.title('Total N2O Emissions Over Time',fontsize=15)
plt.legend()
plt.savefig('Total N2O Emissions Over Time.png', dpi=300, bbox_inches='tight')
plt.show()

#for Nitrous Oxide N2O
```

```python
plt.figure(figsize=(15,8))
plt.plot(sweden['year'],sweden['total_ghg'],color='green',label='Sweden',marker='d')
plt.plot(denmark['year'],denmark['total_ghg'],color='red',label='Denmark', marker='d')
plt.plot(iceland['year'],iceland['total_ghg'],color='magenta',label='Icleand', marker='d')
plt.plot(norway['year'],norway['total_ghg'],color='blue',label='Norway', marker='d')
plt.plot(finland['year'],finland['total_ghg'],color='orange',label='Finland', marker='d')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Total GHGs mT",fontsize=15)
plt.title('Total GHGs Emissions Over Time',fontsize=15)
plt.legend()
plt.savefig('Total GHGs Emissions Over Time.png', dpi=300, bbox_inches='tight')
plt.show()
#for Total Greenhouse Gas Emissions
```

```python
co2_clust =all_countries_sectors_co2[['Agriculture','Land-use change and forestry','Waste','Industry','Manufacturing and construction','Transport','Electricity and heat','Buildings','Fugitive emissions','Other fuel combustion','Aviation and shipping']].copy()
#creating copy for scaling
```

```python
co2_clust.dtypes
#checking data types
```

```python
scale=MinMaxScaler()
scaled_data_co2 =scale.fit_transform(co2_clust)
scaled_data_co2
#scaling the data
```

```python
distortions = []
K = range(1,6)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(scaled_data_co2)
    distortions.append(kmeanModel.inertia_)
#applying kMeans
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
kn = KneeLocator(K, distortions, curve='convex', direction='decreasing')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dotted', color='red')
plt.xlabel('Number of K', fontsize=15)
plt.ylabel('Distortion', fontsize=15)
plt.grid(False)
plt.title('Elbow Method')
plt.savefig('elbow.png', dpi=300, bbox_inches='tight')
plt.show()
#plotting the elbow for appropriate no. of K or clusters
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(scaled_data_co2)
#here we take the corresponding no. of clusters
```

```python
plt.figure(figsize=(16,8))
plt.scatter(scaled_data_co2[:,0],scaled_data_co2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.title('Formation of Clusters')
plt.savefig('elbow.png', dpi=300, bbox_inches='tight')
#checking formation of different clusters
```

```python
data_reduced = pd.DataFrame(scaled_data_co2)
all_countries_sectors_co2['Cluster'] = kmeans.labels_.tolist()
all_countries_sectors_co2
#adding the cluster to the row in initial dataframe
```

```python
all_countries_sectors_co2[['Cluster','Agriculture','Land-use change and forestry','Waste','Industry','Manufacturing and construction','Transport','Electricity and heat','Buildings','Fugitive emissions','Other fuel combustion','Aviation and shipping']].groupby('Cluster').mean().head()
#grouping features by using clusters to judge the difference in attributes
```

```python
nordic_clust =nordic_total_emissions_sectors[['Agriculture','Land-use change and forestry','Waste','Industry','Manufacturing and construction','Transport','Electricity and heat','Buildings','Fugitive emissions','Other fuel combustion','Aviation and shipping']].copy()
#creating copy for scaling
```

```python
nordic_clust.dtypes
#checking the data types
```

```python
scale=MinMaxScaler()
scaled_data_nordic=scale.fit_transform(nordic_clust)
scaled_data_nordic
#scaling the data
```

```python
distortions_nordic = []
K = range(1,6)
for k in K:
    kmeanModel_nordic = KMeans(n_clusters=k)
    kmeanModel_nordic.fit(scaled_data_nordic)
    distortions_nordic.append(kmeanModel_nordic.inertia_)
#applying kMeans
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions_nordic, 'bx-')
kn = KneeLocator(K, distortions_nordic, curve='convex', direction='decreasing')
plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dotted', color='red')
plt.xlabel('Number of K', fontsize=15)
plt.ylabel('Distortion', fontsize=15)
plt.grid(False)
plt.title('Elbow Method')
plt.savefig('elbow.png', dpi=300, bbox_inches='tight')
plt.show()
#plotting the elbow for appropriate no. of K or clusters
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(scaled_data_nordic)
#here we take the corresponding no. of clusters
```

```python
plt.figure(figsize=(16,8))
plt.scatter(scaled_data_nordic[:,0],scaled_data_nordic[:,1], c=kmeans.labels_, cmap='rainbow')
#checking the formation of different clusters
```

```python
data_reduced_nordic= pd.DataFrame(scaled_data_nordic)
all_countries_summed['Cluster'] = kmeans.labels_.tolist()
all_countries_summed
#adding the cluster to the row in initial dataframe
```
